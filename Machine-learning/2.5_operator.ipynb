1. Arithmatic Operator

```python
num_1=10
num_2=20

#addition
sum=num_1+num_2
print("Addition = ",sum)

#subtraction
diff=num_1-num_2
print("difference = ",diff)

#multiplication

pro=num_1*num_2
print("Product = ",pro)

#division
quo=num_1/num_2
print("quotient = ",quo)

#exponent
exp=num_1**num_2
print("Exponent = ",exp)

#Modulus
mod=num_1%num_2
print("remainder = ",mod)
```

3. Comparison operator


```python
a=5
b=10


print(a==b)  #equal to
print(a!=b)  # no equal to
print(a>b)   #greater than
print(a<b)   #less than
print(a>=b)  #grater than equal to
print(a<=b)   #less than equal to
```

4. Logical operator

```python
a=10

print(a>20 and a>5)
print(a>20 or  a>5)
print(not(a<8 and a<100))
```
