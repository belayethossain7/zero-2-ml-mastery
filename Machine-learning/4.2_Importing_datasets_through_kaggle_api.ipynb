```python
# extracting the compressed dataset

from zipfile import ZipFile        # zipfile libarary theke Zipfile function k call korlam
dataset ='/content/LANL-Earthquake-Prediction.zip'

with ZipFile(dataset,'r') as zip:                           # with keyword or fuction acctually used for OPEN a file then calling function ZipFile() and first parameter is dataset,, jeta te ami path rekh diyesi,,ai dataset variable
  zip.extractall()                                                          # a zip file ar path rekhe diyesi ,,oitai first parameter
  print('the dataset is extracted')                                                          # secound parameter is r , r means read , rad the file
      # extractall function ta oi file ta k etract korbe
  print('the dataset is extracted')
```
