```python
#importing libararies

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as ap
```

Note: seaborn has some builtin data set

```python
#total bill vs tip data set

tip=sns.load_dataset('tips') # akhane sns name seaborn a built in data set ase sekhan theke tips name data set ta akhane load koresi and tip namer variable rekhe diysi
                             # and ai data set ta import hobe pandas data frame ar moto
                             # tai ata jodi head diye print kori easly print hobe
tip.head()
```

```python
# setting a theme for the plots

sns.set_theme() # ai theme fuction dara plot ar background a theme set hoye jabe,,box box theme jeta onk sundor vabe plot represent korbe and sundor dekhabe data ta plot kore visualize korbe
                # aita shudhu akbar code a run korlei hobe , tai sob plot a theme use hobe
                # defult akta theme set kore dibe
```

```python
# visualize the data , akhane ami easly table or data set ta easyly visualize korbo seaborn liabrary diye

sns.relplot(data=tip,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size') # relplot mane relative plot and ata relative plot kore dibe ai data set ta k, ami k vabe dibo ,se vabe print or plot kore dibe
                                                                                                # data te tip rakhsi karon ata k plot korbo
                                                                                                # x axies a total bill and y axies a tip rekhe diyesi . agula total bill and tip holo colom , sob e asole column
                                                                                                # col mane column , column a time rakhsi thats means time j koita individual value ase oi koita column ba plot korbe ,akhane 2 ta korbe karon time a lunch and dinnner ase , only ai duita
                                                                                                # hue holo different color dibe , akhane smoker column diye dewa ase mane smoker kare j koita individual man dewa ase oikota differen color diye ,invidual man prokas korbe ,akhane 2 ta ase tai 2 ta individual color dibe , akta smoker and arekta non smoker
                                                                                                # style same hue ar motoi ata shudhu different shape dibe , individual maner jonno
                                                                                                # size ta different type dot shape dibe , size j variable thakbe , oi variable ar individual value gula k different shape dibe j mn akhane size variable e dewa and aita te 6 pojonto ase ,1 ar jonno choto dot , 2 ar jonno ar aktu boro , 6 sob cheye boro,,avabe protekta value diye different dot size dibe

```

another data set


```python
#boston house dataset
#house price dataset

from sklearn.datasets import fetch_openml   #sklearn liberary theke ai boston data set import korlam

housing = fetch_openml(data_id=531, as_frame=True) # boston dataset ta akta variable  rakhsi ;

house_df=pd.DataFrame(housing.data, columns=housing.feature_names) # housing to data set ta to numpy array type silo ,,, tai avabe dataframe convert a koresi and akta variable rakhsi ,house_df variable a rekhe diyesi
house_df['Price']=housing.target    # house_df dataframe akta price name ar column add koresi , oi price column a housing dataset ar target name ar numpy array ar value gula diye disi

house_df.head()     # 5 raw print
```

important plot

Distribution plot



```python
sns.displot(house_df['Price'])   # akhane house_df data frame ar Price column ar distribution plot kore dibe
                                 # x axies a thake price , and y axies a thakbe count
                                 # that means kon kon price a koto jon manus kinese house sevabe plot or bar dibe
```

Vary vary important plot

Correlatin

*positive correlation(one column value increase and another column value also increase)

*negative correlation(one column value increase and another column value decrease)

Correlation plot kore heat map diye

*Heat Map


```python
# make correlation in housing_df dataframe
correlation=house_df.corr()  # house_df detaframe ta k fully correlation koresi akhane ,,and correlate kore akta variable rekhe diyesi,,variable ta correlation
```

```python
# Constructing Heat map

plt.figure(figsize=(10,10))     # matplotlib diye akta empty figure niyesi jar fig size diye diyesi


sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot= True, annot_kws={'size':8},cmap='Blues')

# correlation: heat map a j correlation korbo ,correlation diye diyese cz correlation a house_df ta correlation correlation kora ase
# cbar: cbar mane color bar akhne j dan pase akta color bar dewa ase ata true bole disi ,, mane color bar show korte bolsi , ai color diye buja jabe minus man color gulo negative correlation j gula light blue and positive correlation holo plus man gula j gula dark blue
# square : plot square shape cheyesi tai true diyesi , and tai square shape diyese
# fmt : fmt mane hosse dosomik ar por koto ghor pjonto value print korate chassi.,,akhane 1 ase tarmane 1 ghor dekkhase
# annot: correlation j koresi , column to column correlation vlaue gula annot korbe ba print korbe
# annot_kws: oi annotation value ar font size ata
# cmap : cmap color dite hoi,, kon color a plot korbe
```
