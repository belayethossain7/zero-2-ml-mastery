This is Highly Imblanced Dataset


0 ---> Legit Transactions

1 ---> fraudulent Transactions

```python
# separating the legit and frudulent transaction

legit = credit_card_data[credit_card_data.Class == 0.0]
fraud = credit_card_data[credit_card_data.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of legit & fraudulent transactions


Number of fraudulent transactions ---> 88

```python
#sampling
legit_sample = legit.sample(n=88)
```

```python
print(legit_sample.shape)
```

Concatenate the two datafrmaes

```python
new_dataset = pd.concat([legit_sample,fraud], axis = 0) # axis = 0 means raw  wise concatanation work
```
