Matplotlib

useful for making plots and graph



```python
#importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

ploting the data

```python
#sine wave

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
# parabola

x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'r+') # ai r+ mane hosse red color dibe and plot korbe + symbol diye
plt.xlabel('x')
plt.ylabel('y')
plt.title('parbola')
plt.show()
```

```python
plt.plot(x,y,'g.') # ai g. diye bujaise j g te green and . dara dot line bujaise mane green color dot line plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'rx') # red color x type plot korbe
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-') # g- dara green color normal line a plot korbe and
plt.plot(x,np.cos(x),'r--')# r-- dara red color hifane hifane line print korbe
plt.xlabel('x')
plt.ylabel('y')
plt.title('2 parabola graph')
plt.show()
```

Bar plot

```python
fig=plt.figure() # ata diye akta faka figure newa hoyese
ax=fig.add_axes([0,0,1,1]) #akhane plot ba ractangle ar size deya hoyese ba plot make kora hoyese ,,0,0 holo mulbindu or orgin and 1,1 holo ar hight and width, tarme ai plot ba ractangle ar origin holo 0,0 and hight and width 1,1
language=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(language,people)  # akhane oi x y ar value niye ai function diye bar diagrma make korte bolsi
plt.xlabel('Language')
plt.ylabel('Number of People')
plt.title('Language used by visitor for Communication')
plt.show()
```

Pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
language=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
plt.title('Language use by visitor for communication')
ax.pie(people,labels=language,autopct='%1.1f%%') # akhene x okkhe people thakbe karon koto parent people parcent hisebe alada korbe ..cz pie chart to 360 degree oi onujaiye people songkkha diye parcent korb ,,
                                                 # r label a language rakhsi karon hosse koto parcent kon laguage label set kore ai y okkho,, r autopct hosse percent hula j korbo dosomik ar por koto ghor print korbe
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2= plt.figure()
ax=fig2.add_axes([0,0,1,1])
plt.xlabel('angle')
plt.ylabel('sine and cosine value')
plt.title("Sine and Cosine wave")
ax.scatter(x,y,color='g') # scatter fuction diye scatter plot kore dibe and dot gula green color ar hobe
ax.scatter(x,z,color='r') # scatter fuction diye scatter kore dibe and color section dite hobe cz dot gula kon color hobe ta bole dite hobe akhen r dara red color bujaise
plt.show()
```

3d scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues') # 3d te 3 ta axis lage , x,y and z, c ar man dara amra anothe axies z axies kei bujailam and , cmap dara color bujailam
plt.show()
```

```python
fig4=plt.figure()
ax=plt.axes(projection='3d')
z= 20* np.random.random(100)
x=np.sin(z)
y=np.cos(z)
plt.xlabel('x')
plt.ylabel('y')

plt.title('Example of 3d line plot')
ax.scatter(x,y,z,c=z,cmap='Reds')

plt.show()
```
